```python
f = open("../data/20150101_20151231_nature_index.txt", "r")
lines = f.readlines()
f.close()
```

```python
absolute_count = []
fractional_count = []
weighted_fractional_count = [] 
names = []
uni_names = ['Los Andes University','University of Antioquia (UdeA)',
           'National University of Colombia (UNAL)',
             'National Autonomous University of Mexico (UNAM)',
             'Technion-Israel Institute of Technology (IIT)']

colors = ['blue', 'red', 'green', 'yellow', 'magenta']
uni_ID = range(len(uni_names))
print uni_ID

for line in lines:
    if line[0]!='#':
        words = line.split()
        #print words
        ac = float(words[-3])
        fc = float(words[-2])
        wfc = float(words[-1])
        name = " ".join(words[1:-3])
        absolute_count.append(ac)
        fractional_count.append(fc)
        weighted_fractional_count.append(wfc)
        names.append(name)
```

```python
scatter(log10(absolute_count), log10(weighted_fractional_count))
```

```python
n_uni = len(uni_names)
for i in range(n_uni):
    uni_ID[i] = names.index(uni_names[i])
print uni_ID
absolute_count = np.array(absolute_count)
weighted_fractional_count = np.array(weighted_fractional_count)
fractional_count = np.array(fractional_count)
```

```python
def plot_xy(x,y ,loc=2):
    scatter(x, y, color='black', s=10.0, alpha=0.2)
    for i in range(n_uni):
        scatter(x[uni_ID[i]], y[uni_ID[i]], s=100, label=uni_names[i], color=colors[i])
    

    xscale('log')
    yscale('log')
    legend(loc=loc, fontsize=9)
```

```python
x = absolute_count
y = weighted_fractional_count
plot_xy(x,y)
```

```python
x = absolute_count
y = absolute_count/weighted_fractional_count
plot_xy(x,y)
```

```python
x = fractional_count
y = absolute_count/weighted_fractional_count
plot_xy(x,y, loc=4)
```
